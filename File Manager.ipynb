```python
'''
This is a Python based project on File Manager
made by Deepesha Burse
as a personal project.
24-July-2021

https://github.com/deepeshaburse/Python-File-Manager
'''

#Importing all required modules
import sys
import os
import shutil
import send2trash
```

```python
#Takes in all the drives in the computer in a list named 'drives'.
drives = [chr(x) + ':' for x in range(65,90) if os.path.exists(chr(x) + ':')]
```

```python
#Function to display all the files and folders in the current directory.
def listDir():
    listOfDir = os.listdir(os.getcwd())
    for x in listOfDir:
        print(x)
```

```python
print("Welcome to the Python File Manager")
while True:
    print("1. Open File/Folder\n2. Rename\n3. Move and Paste\n4. Copy and Paste\n5. Delete\n")
    choice = input("Enter your choice : ")
    if choice == '1':
        #Home Screen
        print("\nQuick Access:\n1. Documents\n2. Videos\n3. Pictures\n4. Downloads\n")
        
        print("\nDrives :")
        for x in range(len(drives)):
            print(str(1+x) + ". " + drives[x])
        print("\n")    
        while True:
            inp = input("Enter your choice : ")
            if inp == '1':
                path = 'C:\\Users\\$USERNAME\\Documents'
                os.chdir(os.path.expandvars(path))
                break
            elif inp == '2':
                path = 'C:\\Users\\$USERNAME\\Videos'
                os.chdir(os.path.expandvars(path))
                break
            elif inp == '3':
                path = 'C:\\Users\\$USERNAME\\Pictures'
                os.chdir(os.path.expandvars(path))
                break
            elif inp == '4':
                path = 'C:\\Users\\$USERNAME\\Downloads'
                os.chdir(os.path.expandvars(path))
                break
            elif inp in drives:
                os.chdir(inp + "\\")
                break
            else:
                print("Error\nEnter a valid option/drive name\n")
                
        while True:
            listDir()
            
            print('\n\nEnter "exitManager" to exit the file manager')
            print('\n\nEnter "backManager" to go up one directory')
            print('\n\nEnter "otherFeature" to do another function')
            
            res = input("\nChoose a file/folder : ")
            print("\n")
            
            if res in os.listdir(os.getcwd()):
                if os.path.isfile(res):
                    os.system('"' + res + '"')
                else:
                    os.chdir(res)
            
            #To exit the file manager
            elif res == 'exitManager':
                sys.exit(0)
            
            #To go up one directory
            elif res == 'backManager':
                os.chdir('..')
                
            #To use other features of the manager
            elif res == 'otherFeature':
                break
                
            else:
                print("No folder/file of this name exists.")
                
    elif choice == '2':
        print('You chose to rename')
        print('\nDrives :')
        for x in range(len(drives)):
            print(str(1+x) + ". " + drives[x])
        print("\n")
        
        while True:
            inp = input("Enter your choice : ")
            if inp in drives:
                os.chdir(inp + "\\")
                break
            else:
                print("Error\nEnter a valid drive name.\n")
                
        while True:
            listDir()
            
            print('\n\nEnter "exitManager" to exit the file manager')
            print('\n\nEnter "backManager" to go up one directory')
            print('\n\nEnter "renameDir" to rename this directory')
            print('\n\nEnter "otherFeature" to do another function')
            
            res = input("\nChoose a file to rename : ")
            print("\n") 
            
            if res in os.listdir(os.getcwd()):
                if os.path.isfile(res):
                    
                    newname = input("Enter a new name: ")
                    originalDir = res
                    newDir = os.getcwd() + '\\' + newname
                    shutil.move(originalDir, newDir)
                    
                else:
                    os.chdir(res)
                
            elif res == 'exitManager':
                sys.exit(0)
                
            elif res == 'backManager':
                os.chdir('..')
            
            elif res == 'renameDir':
                newname = input("Enter a new name: ")
                originalDir = os.getcwd()
                os.chdir('..')
                newDir = os.getcwd() + '\\' + newname
                shutil.move(originalDir, newDir)
                
            elif res == 'otherFeature':
                break
                
            else:
                print("No folder/file of this name exists.")
                
    elif choice == '3':
        print('You chose to move')
        print('\nDrives :')
        for x in range(len(drives)):
            print(str(1+x) + '. ' + drives[x])
        
        while True:
            inp = input("Enter your choice : ")
            if inp in drives:
                os.chdir(inp + "\\")
                break
            else:
                print("Error\nEnter a valid drive name.\n")
                
        while True:
            listDir()
            
            print('\n\nEnter "exitManager" to exit the file manager')
            print('\n\nEnter "backManager" to go up one directory')
            print('\n\nEnter "moveDir" to move this directory')
            print('\n\nEnter "otherFeature" to do another function')
            
            res = input("\nChoose a file to move : ")
            print("\n") 
            
            if res in os.listdir(os.getcwd()):
                
                if os.path.isfile(res):
                    originalpath = os.getcwd() + "\\" + res
                    print("\nMove " + res + " to a desired location")
                    
                    while True:
                        for x in range(len(drives)):
                            print(str(1+x) + ". " + drives[x])
                            
                        inp2 = input("\nEnter your choice: ")
                        
                        if inp2 in drives:
                            os.chdir(inp2 + '\\')
                            break
                            
                        else:
                            print("Error\nEnter a valid drive name.\n")
                            
                    while True:
                        listDir()
                        
                        print('Type "pasteFile" to paste this file in the current directory')
                        
                        res2 = input("\nChoose a file to move: ")
                        print("\n")
                        
                        if res2 in os.listdir(os.getcwd()):
                            if os.path.isfile(res):
                                print("You can't choose this file.\nPlease choose a folder.")
                            else:
                                os.chdir(res2)
                                
                        elif res2 == 'pasteFile':
                            shutil.move(originalpath, os.getcwd())
                            break
                            
                else:
                    os.chdir(res)
                    
            elif res == 'exitManager':
                sys.exit(0)
                
            elif res == 'backManager':
                os.chdir('..')
                
            elif res == 'otherFeature':
                break
                
            elif res == 'moveDir':
                originalpath = os.getcwd()
                
                print("Moving the current directory")
                while True:
                    for x in range(len(drives)):
                        print(str(1 + x) + '. ' + drives[x])

                    inp2 = input("\nEnter your Choice: ")

                    if inp2 in drives:
                        os.chdir(inp2 + '\\')
                        break
                    else:
                        print('Error\nEnter a correct drive name.\n')

                while True:
                    listDir()

                    print('\nType "pasteFolder" to paste this folder in current directory')

                    res2 = input('\nChoose a folder to open: ')
                    print('\n')

                    if res2 in os.listdir(os.getcwd()):
                        if os.path.isfile(res):
                            print("You can't choose a file.\nPlease choose a folder.")
                        else:
                            os.chdir(res2)

                    elif res2 == 'pasteFolder':
                        shutil.move(originalpath, os.getcwd())
                        break
            
            else:
                print("No such file/folder exists.")
                
    elif choice == '4':
        print("You chose to copy")
        print("\nDrives :")
        for x in range(len(drives)):
            print(str(1+x) + ". " + drives[x])
            
        while True:
            inp = input("\nEnter your choice : ")
            
            if inp in drives:
                os.chdir(inp + "\\")
                break
                
            else:
                print("Error\nEnter a valid drive name.\n")
                
        while True:
            listDir()
            
            print('\n\nEnter "exitManager" to exit the file manager')
            print('\n\nEnter "backManager" to go up one directory')
            print('\n\nEnter "copyDir" to copy this directory')
            print('\n\nEnter "otherFeature" to do another function')
            
            res = input("\nChoose a file to move : ")
            print("\n") 
            
            if res in os.listdir(os.getcwd()):
                
                if os.path.isfile(res):
                    originalpath = os.getcwd() + "\\" + res
                    print("Move " + res + " to a desired location.")
                    
                    while True:
                        for x in range(len(drives)):
                            print(str(1+x) + ". " + drives[x])
                            
                        inp2 = input("\nEnter your choice")
                        
                        if inp2 in drives:
                            os.chdir(inp2 + "\\")
                            break
                            
                        else:
                            print("Error\nEnter a valid drive name.\n")
                            
                    while True:
                        listDir()
                        
                        print('Type "pasteFile" to paste this file in the current directory')
                        
                        res2 = input("\nChoose a file to move: ")
                        print("\n")
                        
                        if res2 in os.listdir(os.getcwd()):
                            if os.path.isfile(res):
                                print("You can't choose this file.\nPlease choose a folder.")
                            else:
                                os.chdir(res2)
                                
                        elif res2 == 'pasteFile':
                            shutil.copy(originalpath, os.getcwd())
                            break
                            
                else:
                    os.chdir(res)
                    
            elif res == 'exitManager':
                sys.exit(0)
                
            elif res == 'backManager':
                os.chdir('..')
                
            elif res == 'otherFeature':
                break
            
            elif res == 'copyDir':
                originalpath = os.getcwd()
                
                print("Copying the current directory")
                while True:
                    for x in range(len(drives)):
                        print(str(1+x) + ". " + drives[x])
                        
                    inp2 = input("\nEnter your choice")
                    
                    if inp2 in drives:
                        os.chdir(inp2 + "\\")
                        break
                    
                    else:
                        print("Error\nEnter a valid drive name.\n")
                        
                while True:
                    listDir()
                    
                    print('\nType "pasteFile" to copy this file in current directory.')
                    
                    res2 = input("\nChoose a folder to open")
                    print("\n")
                    
                    if res2 in os.listdir(os.getcwd()):
                        if os.path.isfile(res):
                            print("You can't choose this file.\nPlease choose another folder.")
                            
                        else:
                            os.chdir(res2)
                            
                    elif res2 == 'pasteFile':
                        print(originalpath)
                        foldername =  originalpath.split('\\')[-1]
                        folderdirectory = os.getcwd() + '\\' + foldername
                        shutil.copytree(originalpath, folderdirectory)
                        break
            
            else:
                print("No file/folder of this name exists.")
                
    elif choice == '5':
        while True:
            print("\n1. Permanently Delete \n2. Recycle Bin")
            query = input("Would you like to permanently delete the file or send it to the Recycle Bin? ")
            
            if query == '1':
                print("You chose to permanently delete files/folders.\n")
                print("\nDrives :")
                for x in range(len(drives)):
                    print(str(1+x) + ". "  + drives[x])
                
                while True:
                    inp = input("Enter your choice")
                    
                    if inp in drives:
                        os.chdir(inp + "\\")
                        break
                        
                    else:
                        print("Error\nEnter a valid drive name.\n")
                        
                while True:
                    listDir()
                    
                    print('\n\nEnter "exitManager" to exit the file manager')
                    print('\n\nEnter "backManager" to go up one directory')
                    print('\n\nEnter "delDir" to permanently delete this directory')
                    print('\n\nEnter "otherFeature" to do another function')
            
                    res = input("\nChoose a file to delete : ")
                    print("\n")
                    
                    if res in os.listdir(os.getcwd()):
                        if os.path.isfile(res):
                            
                            #Warning message 
                            print("Are you sure you want to permanently delete this file/folder? (YES/NO)")
                            
                            ans = input("Yes or no ")
                            
                            if ans.lower() == 'yes' or 'y':
                                os.unlink(res)
                                
                        else:
                            os.chdir(res)
                            
                    elif res == 'exitManager':
                        sys.exit(0)
                
                    elif res == 'backManager':
                        os.chdir('..') 
                    
                    elif res == 'otherFeature':
                        break
                        
                    elif res == 'delDir':
                        
                        #Warning message
                        print("Are you sure you want to permanently delete this file/folder? (YES/NO)")
                            
                        ans = input("Yes or no ")
                            
                        if ans.lower() == 'yes' or 'y':
                            path = os.getcwd()
                            os.chdir('..')
                            shutil.rmtree(path)
                                
                    else:
                        print("No file/folder of this name exists.")
                        
            elif query == '2':
                print("You chose to temporarily delete the files/folders.")
                print("\nDrives :")
                for x in range(len(drives)):
                    print(str(1+x) + ". " + drives[x])
                    
                while True:
                    inp = input("\nEnter your choice")
                    
                    if inp in drives:
                        os.chdir(inp + "\\")
                        break
                        
                    else:
                        print("Error\nEnter a valid drive name.\n")
                        
                while True:
                    listDir()
                    
                    print('\n\nEnter "exitManager" to exit the file manager')
                    print('\n\nEnter "backManager" to go up one directory')
                    print('\n\nEnter "delDir" to temporarily delete this directory')
                    print('\n\nEnter "otherFeature" to do another function')
            
                    res = input("\nChoose a file to delete : ")
                    print("\n")
                    
                    if res in os.listdir(os.getcwd()):
                        if os.path.isfile(res):
                            
                            #Warning message
                            print("Are you sure you want to send this file/folder to the recycle bin? (YES/NO)")
                            ans = input("Yes or no ")
                            
                            if ans.lower() == 'yes' or 'y':
                                send2trash.send2trash(res)
                                
                        else:
                            os.chdir(res)
                            
                    elif res == 'exitManager':
                        sys.exit(0)
                
                    elif res == 'backManager':
                        os.chdir('..') 
                        
                    elif res == 'otherFeature':
                        break
                        
                    elif res == 'delDir':
                        
                        #Warning message
                        print("Are you sure you want to temporarily delete this file/folder? (YES/NO)")
                            
                        ans = input("Yes or no ")
                            
                        if ans.lower() == 'yes' or 'y':
                            path = os.getcwd()
                            os.chdir('..')
                            send2trash.send2trash(path)
                                
                    else:
                        print("No file/folder of this name exists.") 
    else:
        print("You chose the wrong number")
```
